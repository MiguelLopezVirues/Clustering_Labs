```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import xicorpy as xicor
```

```python
x = np.linspace(-12, 12, 500)
y = np.exp(-((x - 3)**2) / 4) - np.exp(-((x + 3)**2) / 4) + np.random.normal(0, 0.01, size=len(x))

plt.scatter(x, y)
plt.title("Non-Linear Non-Monotonic Relationship")
plt.show()

xi = xicor.compute_xi_correlation(x, y)

xi_value, p_value = xicor.compute_xi_correlation(x, y, get_p_values=True)

print(f"Pearson's correlation is {np.corrcoef(x,y)[0][1]}")
spearman = pd.DataFrame({"x":x,"y":y}).corr(method="spearman").iloc[1,0]
print(f"Spearman's correlation is {spearman}")
print(f"Chatterjee's Xi Correlation: {xi_value}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.linspace(0, 9 * np.pi, 1000)
y = np.sin(x) + np.random.normal(0, 0.1, size=len(x))  # Centered around 0

# Plot the data
plt.scatter(x, y)
plt.axhline(0, color='red', linestyle='--', label='y = 0')
plt.title("Sine Function Centered Around 0")
plt.legend()
plt.show()

xi = xicor.compute_xi_correlation(x, y)

xi_value, p_value = xicor.compute_xi_correlation(x, y, get_p_values=True)

print(f"Pearson's correlation is {np.corrcoef(x,y)[0][1]}")
spearman = pd.DataFrame({"x":x,"y":y}).corr(method="spearman").iloc[1,0]
print(f"Spearman's correlation is {spearman}")
print(f"Chatterjee's Xi Correlation: {xi_value}")
```

```python
x = np.linspace(0, 10, 100)
y = np.exp(x / 1.5) + np.random.normal(0, 0.5, size=len(x))

plt.scatter(x, y)
plt.title("Exponential Relationship")
plt.show()

xi = xicor.compute_xi_correlation(x, y)

xi_value, p_value = xicor.compute_xi_correlation(x, y, get_p_values=True)

print(f"Pearson's correlation is {np.corrcoef(x,y)[0][1]}")
spearman = pd.DataFrame({"x":x,"y":y}).corr(method="spearman").iloc[1,0]
print(f"Spearman's correlation is {spearman}")
print(f"Chatterjee's Xi Correlation: {xi_value}")
```
